# Data Cleaning

Cleaning of Data

```python
import pandas as pd
```

```python
df = pd.read_excel(r"/content/Dataset -  Assignment 1.xlsx")
df
```

```python
# Remove duplicate
df = df.drop_duplicates()
df
```

```python
# Removing time data from date Column
df['order_date'] = pd.to_datetime(df['order_date']).dt.date
df.head()
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Droping rows with missing values
df = df.dropna()
```

```python
# Renameing columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# Converting text columns to lowercase
df['product_name'] = df['product_name'].str.lower()
```

```python
# Removing outliers in 'final_ak_price'
Q1 = df['final_ak_price'].quantile(0.25)
Q3 = df['final_ak_price'].quantile(0.75)
IQR = Q3 - Q1

# Defining a reasonable range for 'final_ak_price'
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering out outliers
df = df[(df['final_ak_price'] >= lower_bound) & (df['final_ak_price'] <= upper_bound)]
```

```python
# Calculating the total order value for each customer
df['total_order_value'] = df['final_ak_price'] * df['quantity']
customer_data = df.groupby('customer_id')['total_order_value'].sum().reset_index()
```

```python
# Identifing the top 10 customer IDs based on total order value
top_customers = customer_data.sort_values(by='total_order_value', ascending=False).head(10)
print(top_customers)
```

```python
# Saveing the final cleaned DataFrame to a CSV file
df.to_csv('/content/final_cleaned_dataset.csv', index=False)

# Downloading the file
from google.colab import files
files.download('/content/final_cleaned_dataset.csv')
```

# SQL Lite Import

```python
import sqlite3

# Loading the CSV file into a DataFrame
df = pd.read_csv('/content/final_cleaned_dataset.csv')

# Connecting to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Loading the DataFrame into the SQLite database
df.to_sql('orders', conn, index=False, if_exists='replace')
```

# SQL Questions

# Question 1

Average order value

```python
# Connecting to an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Loading the DataFrame into the SQLite database
df.to_sql('orders', conn, index=False, if_exists='replace')

# Defining the SQL query to calculate the overall average order value
query = """
SELECT
    AVG(total_order_value) AS overall_average_order_value
FROM
    orders;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

Total number of orders

```python
query = """
SELECT
    COUNT(order_id) AS total_orders
FROM
    orders;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

First Order date

```python
query = """
SELECT
    MIN(order_date) AS first_order_date
FROM
    orders;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

Last order date

```python
query = """
SELECT
    MAX(order_date) AS last_order_date
FROM
    orders;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

First Order Value

```python
query = """
SELECT
    total_order_value AS first_order_value
FROM
    orders
ORDER BY
    order_date ASC
LIMIT 1;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

Last Order Value

```python
query = """
SELECT
    total_order_value AS last_order_value
FROM
    orders
ORDER BY
    order_date DESC
LIMIT 1;
"""

# Executein the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df)
```

# Question 2

Top 20 Products in Our Popular Product Category

```python
query = """
SELECT
    sku_id,
    product_name,
    SUM(total_order_value) AS total_sales_value
FROM
    orders
GROUP BY
    sku_id, product_name
ORDER BY
    total_sales_value DESC
LIMIT 20;
"""

# Executeing the query and load the result into a DataFrame
top_products_df = pd.read_sql_query(query, conn)


print(top_products_df)

```

# Question 3

Top 3 articles for each customer based on placed_gmv

```python
query = """
SELECT
    product_name,
    SUM(placed_gmv) AS total_gmv
FROM
    orders
GROUP BY
    product_name
ORDER BY
    total_gmv DESC
LIMIT 3;
"""

# Executeing the query and load the result into a DataFrame
result_df = pd.read_sql_query(query, conn)


print(result_df[['product_name']])
```
